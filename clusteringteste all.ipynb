```python
# Boucle pour chaque algorithme de clustering
for algorithm in clustering_algorithms:
    algorithm_name = str(algorithm)

    for score_name, score_func in score_funcs.items():
        scores = []
        figures = []

        for num_cluster in num_clusters:
            if algorithm_name == 'DBSCAN':
                if num_cluster < 2:
                    continue  # DBSCAN ne nécessite pas de nombre de clusters défini
                dbscan = DBSCAN(eps=0.5, min_samples=5)
                cluster_labels = dbscan.fit_predict(reduced_data_3d)
            else:
                cluster_labels = algorithm.fit_predict(reduced_data_3d)

            if len(set(cluster_labels)) < 2:
                continue  # Évitez de calculer les scores pour un petit nombre de clusters

            # Calculer le score
            score = score_func(reduced_data_3d, cluster_labels)
            scores.append(score)

            # Créez un graphique 3D avec Plotly
            scatter = go.Scatter3d(
                x=reduced_data_3d[:, 0],
                y=reduced_data_3d[:, 1],
                z=reduced_data_3d[:, 2],
                mode='markers',
                marker=dict(size=5, opacity=0.7, color=cluster_labels),
                name=f'Clusters: {num_cluster}, Score: {score:.2f}'
            )
            figures.append(scatter)

        # Afficher le graphique 3D
        fig = go.Figure(data=figures)
        fig.update_layout(
            title=f'{algorithm_name} - {score_name}',
            scene=dict(
                xaxis_title='Component 1',
                yaxis_title='Component 2',
                zaxis_title='Component 3'
            )
        )
        fig.show()

        # Afficher les scores pour chaque mesure
        if scores:
            plt.figure(figsize=(10, 5))
            plt.plot(num_clusters, scores, marker='o', linestyle='-')
            plt.title(f'{algorithm_name} - {score_name} vs Number of Clusters')
            plt.xlabel('Number of Clusters')
            plt.ylabel(score_name)
            plt.grid(True)
            plt.show()
```
